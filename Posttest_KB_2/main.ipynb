##### Tampilan Dataset

```python
import pandas as pd
import numpy as np
from prettytable import PrettyTable

df = pd.read_csv("./dataset/vgsales.csv")

df.head(10)
```

##### Ketentuan Point 1: Pengambilan Data
`total_records`: `df.shape` mengembalikan total baris dan total kolom, maka kita akan mengambil data pertamanya

`attributes`: menggunakan `df.columns.values` mengembalikan bentuk numpy array dari `df.columns` 

`object_attributes`: menggunakan `select_dtypes` untuk mengambil baris dengan kolom yang memiliki data type yang diminta, kemudian ambil kolomnya saja

`number_attributes`: menggunakan `select_dtypes` untuk mengambil baris dengan kolom yang memiliki data type yang diminta, kemudian ambil kolomnya saja

`null_attribtues`: `df.isnull().any()` mengembailkan list kolom dan value boolean tentang apakah ada data yang kosong atau tidak, kita akan mengambil yang bernilai `True` dan mengambil nama kolomnya

```python
info = {
  'total_records': 0,
  'attributes': [],
  'object_attributes': 0,
  'number_attributes': 0,
  'null_attributes': 0
}

info['total_records'] = df.shape[0]
info['attributes'] = list(df.columns.values)
info['null_attributes'] = list(df.loc[:, df.isnull().any()].columns)
info['object_attributes'] = list(df.select_dtypes(['object']).columns)
info['number_attributes'] = list(df.select_dtypes(['float64', 'int64']).columns)
```

#### Ketentuan Point 2: Penampilan Data

```python
table = PrettyTable(["Data", "Hasil"])
table.align = "l"

table.add_row(["Jumlah Record", info['total_records']])
table.add_row(["Jumlah Attribute Attribute", f"{', '.join(info['attributes'])} ({ len(info['attributes']) })"])
table.add_row(["Jumlah Attribute Object", f"{', '.join(info['object_attributes'])} ({ len(info['object_attributes']) })"])
table.add_row(["Jumlah Attribute Angka", f"{', '.join(info['number_attributes'])} ({ len(info['number_attributes']) })"])
table.add_row(["Jumlah Attribute Null", f"{', '.join(info['null_attributes'])} ({ len(info['null_attributes']) })"])


print(table)
```

#### Ketentuan Point 3: Rata-rata, dll

```python
secondTable = PrettyTable(["Attribute", "Data", "Hasil"])
secondTable.align = "l"

attributes = list(df.select_dtypes(['int64', 'float64']).columns)


for attribute in attributes:
  dfAttr = df[attribute]
  secondTable.add_row([attribute, "Rata-rata", dfAttr.mean() * 100])
  secondTable.add_row(["", "Standar Deviasi", dfAttr.std(numeric_only=True)])
  secondTable.add_row(["", "Nilai Minimum", dfAttr.min()])
  secondTable.add_row(["", "Nilai Maksimum", dfAttr.max()])
  secondTable.add_row(["", "Q1", dfAttr.quantile(.25)])
  secondTable.add_row(["", "Q2", dfAttr.quantile(.50)])
  secondTable.add_row(["", "Q3", dfAttr.quantile(.75)])
  secondTable.add_row(["", "", ""])

print(secondTable)
```
